```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.DataFrame.from_csv('Crunchbase/investments.csv', index_col = None)
```

```python
df = df[['company_name', 'company_permalink', 'investor_name', 'funding_round_type', 'funding_round_code', 'funded_at', 'raised_amount_usd']]
```

```python
df.head()
```

```python
#Creating Tier Dummy Variable
tier_1 = []
tier_2 = []
tier_3 = []
tier_4 = []
tier_5 = []

for investor in df.investor_name:
    
    #Tier 1
    if investor in {'Andreessen Horowitz', 'Khosla Ventures', 'SV Angel', 'Accel', 'New Enterprise Associates', 'Sequoia Capital', 'Venrock', 'First Round', 'Spark Capital', 'Triangle Peak Partners'}:
        tier_1.append(1)
    else:
        tier_1.append(0)
    
    #Tier 2
    if investor in {'Franklin Square Group', 'Kleiner Perkins Caufield & Byers', 'Google Ventures', 'Founders Fund', 'ARCH Venture Partners', 'Bezos Expeditions', 'Lightspeed Venture Partners', 'Thrive Capital', 'General Catalyst Partners', 'Redpoint Ventures'}:
        tier_2.append(1)
    else:
        tier_2.append(0)
        
    #Tier 3
    if investor in {'Atlas Venture', 'QED Investors', 'Greylock Partners', 'Formation 8', 'Bain Capital Ventures', 'Iridium Communications', 'Lerer Hippeau Ventures', 'Collaborative Fund', 'Bessemer Venture Partners', 'Flagship Ventures'}:
        tier_3.append(1)
    else:
        tier_3.append(0)
    
    #Tier 4
    if investor in {'OrbiMed Advisors', 'SherpaVentures', 'Insight Venture Partners', 'IVP', 'True Ventures', 'Greycroft Partners', 'Menlo Ventures', 'Intel Capital', 'Battery Ventures', 'Tiger Global Management'}:
        tier_4.append(1)
    else:
        tier_4.append(0)
        
    #Tier 5
    if investor in {'Deerfield Capital Management', 'Charles River Associates', 'WRF Capital', 'Benchmark', 'FirstMark Capital', 'CrunchFund', 'Canaan Partners', 'Polaris Partners', 'Sofinnova Partners', 'InterWest Partners'}:
        tier_5.append(1)
    else:
        tier_5.append(0)
```

```python
df['tier_1'] = tier_1
df['tier_2'] = tier_2
df['tier_3'] = tier_3
df['tier_4'] = tier_4
df['tier_5'] = tier_5
```

```python
df.to_csv('VC_tiers.csv')
```
